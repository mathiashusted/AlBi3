**Schritte**

- Datensatz in Seurat-Objekt umwandeln -> für QC-Metriken und zum Filtern von Zellen, um ungewollte Zelen zu entfernen

- Normalisierung der Daten -> um Werte miteinander vergleichen zu können

- Feature Selection -> Gene die in bestimmten Zellen hoch- oder herunterreguliert sind

- Skalierung der Daten -> um eine PCA durchzuführen

- Durchführung einer PCA -> um Daten in eine 2D-Abbildung zu projizieren

- Wahl der Anzahl der Hauptkomponenten -> Wie viele Dimensionen soll der Datensatz beinhalten? Suchen eine geeignete Anzahl an Hauptkomponenten für die Analyse mittels Elbow-Plot (auch Scree-Plot)

- Clustering der Zellen + Dimensionsreduktion -> Suchen Zellen, die ähnliche Eigenschaften besitzen (bestimmen Labels anhand ihrer nächsten Nachbarn), und reduzieren den Datensatz auf 2 Dimensionen zur Darstellung

- Suchen Gene, die anders differenziert sind -> Findet positive und negative Markers innerhalb eines Clusters (es ist auch möglich, Cluster untereinander zu vergleichen) und weist den Clustern ihren Markern zu

**Interpretation**

Der Scatterplot am Ende zeigt, dass die Gene anhand ihrer Expression in den Zellen in 9 verschiedene Cluster eingeteilt werden können (Naive CD4 T, CD14+ Mono, Memory CD4 T, B, CD8 T, FCGR3A+ Mono, NK, DC, Platelet). Die Gene im Cluster Platelet streuen am wenigsten, während Cluster wie Memory CD4 und TNaive CD4 T viel stärker streuen.

Die Heatmap zeigt, dass für Cluster 1 (hier laut Deklarierung CD14+ Mono) bestimmte Gene sehr stark voneinander abhängen bzw. stark reguliert werden, während außerhalb des Clusters diese Gene viel weniger exprimiert werden. Analog gilt dies auch für die anderen Cluster mit anderen Genen. Somit kann man Gene auch in ihre Cluster anhand des Heatmaps erkennen.
