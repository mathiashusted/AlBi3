# Aufgabe 1: Moderne DNA-Sequenzierungsmethoden

## 1. Illumina Sequenzierung (Short-Read Sequenzierung)

**Beschreibung:**  
Bei der Illumina-Sequenzierung wird DNA in sehr kurze Stücke geschnitten.
Diese kurzen Stücke werden parallel gelesen. Die Methode ist sehr genau und
wird weltweit am häufigsten eingesetzt.

**Typische Anwendungen:**  
- Ganzgenom-Sequenzierung  
- RNA-Sequenzierung  
- Medizinische und biologische Forschung  

**Sequenziermenge pro Tag:**  
- Moderne Geräte (z. B. NovaSeq): ca. **3–10 Terabasen (Tb) pro Tag**

**Vorteile:**  
- Sehr hohe Genauigkeit  
- Sehr hoher Durchsatz  

**Nachteile:**  
- Nur kurze DNA-Sequenzen (Reads)

**Quelle:**  
- Illumina – NovaSeq Produktseite  
  https://www.illumina.com

---

## 2. PacBio SMRT Sequenzierung (Long-Read Sequenzierung)

**Beschreibung:**  
PacBio liest einzelne DNA-Moleküle in Echtzeit. Die DNA-Sequenzen sind deutlich
länger als bei Illumina und trotzdem sehr genau.

**Typische Anwendungen:**  
- Zusammenbau neuer Genome  
- Analyse komplexer Genomregionen  
- Strukturvarianten

**Sequenziermenge pro Tag:**  
- Je nach Gerät (Sequel II, Revio): ca. **50–400 Gigabasen (Gb) pro Tag**

**Vorteile:**  
- Lange Reads  
- Sehr hohe Genauigkeit  

**Nachteile:**  
- Geringerer Durchsatz als Illumina  
- Hohe Kosten

**Quelle:**  
- Pacific Biosciences – SMRT Sequencing  
  https://www.pacb.com

---

## 3. Oxford Nanopore Sequenzierung

**Beschreibung:**  
Bei dieser Methode wird DNA durch winzige Poren (Nanoporen) geleitet.
Dabei entstehen sehr lange Sequenzen. Die Daten können sofort ausgewertet werden.

**Typische Anwendungen:**  
- Sehr lange DNA-Sequenzen  
- Schnelle Analysen  
- Metagenomik

**Sequenziermenge pro Tag:**  
- Große Geräte (z. B. PromethION): ca. **1–5 Terabasen (Tb) pro Tag**

**Vorteile:**  
- Sehr lange Reads möglich  
- Flexible Laufzeiten  

**Nachteile:**  
- Etwas geringere Genauigkeit als Illumina und PacBio

**Quelle:**  
- Oxford Nanopore Technologies  
  https://nanoporetech.com

---

## Vergleich der Methoden

| Methode    | Read-Länge | Genauigkeit | Durchsatz pro Tag |
|-----------|-----------|-------------|------------------|
| Illumina  | kurz      | sehr hoch   | sehr hoch (Tb)   |
| PacBio    | lang      | sehr hoch   | mittel (Gb)     |
| Nanopore  | sehr lang | mittel–hoch | hoch (Tb)       |

---

## Fazit

- **Illumina**: Beste Wahl für große Datenmengen und hohe Genauigkeit  
- **PacBio**: Ideal für lange und sehr genaue Sequenzen  
- **Nanopore**: Sehr flexibel und gut für extrem lange DNA-Reads

Kurze Erklärung:

- **QNAME**  
  Name des Reads

- **FLAG**  
  Zahlen-Code mit Infos zum Read (z. B. gepaart, Richtung, gemappt)

- **RNAME / CHR**  
  Referenzsequenz bzw. Chromosom

- **POS**  
  Startposition des Reads auf dem Chromosom

- **MAPQ (MQ)**  
  Qualität der Ausrichtung (je höher, desto sicherer)

- **CIGAR**  
  Beschreibt, wie der Read zur Referenz passt  
  (z. B. Matches, Insertionen, Deletionen)

- **RNEXT**  
  Chromosom des Mate-Reads (bei paired-end)

- **PNEXT**  
  Position des Mate-Reads

- **TLEN**  
  Länge des Fragments

- **SEQ**  
  Die gelesene DNA-Sequenz

- **QUAL**  
  Qualitätswerte der Basen
