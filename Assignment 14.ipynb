### Aufgabe 2: GSEA
a) Lesen Sie in dem GSEA paper (https://www.pnas.org/doi/abs/10.1073/pnas.
0506580102) nach, wie die Gene, die zur gesuchten Gruppe und die Gene, die nicht
dazu gehören, in der Summierung gewichtet werden (Step 1 und Appendix des Papers,
es sei p = 1).

b) Warum endet die GSEA Kurve letztlich wieder auf der x-Achse?

c) Was ist MSigDB?

**a)**

Alle Gene werden nach ihrer Assoziation mit dem Phänotypen sortiert. Dann wird für jedes Gen in der Liste die Score erhöht, wenn das Gen zur gesuchten Gruppe gehört bzw. verringert, falls nicht. Die Verringerung ist dabei immer ein konstanter Wert, die Erhöhung um einen Gewichtungsbetrag, der vom Rang des Gens abhängt. Für $p = 1$ ist der Gewichtungsbetrag abhängig vom absoluten Score.

**b)**

Die Gewichte sind so normiert, dass die Summe aller positiven Schritte der Summe negativer Schritte entspricht. Der Enrichtment Score ES ist bloß der maximale Abstand von der x-Achse, den die Laufkurve erreicht.

**c)**

MSigDB steht für Molecular Signature Database und ist ein Katalog von 1325 Genen, die biologisch sinnvoll gruppiert sind (z.B. nach Signalwegen). Die macht sich GSEA zu Nutze um die Gewichtung zu berechnen.
